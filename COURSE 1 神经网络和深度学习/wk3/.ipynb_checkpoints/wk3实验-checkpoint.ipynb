- 初始化随机参数

```python
def initialize_parameters(n_x,n_h,n_y):
    #设置随机种子 
    np.random.seed(2)
    W1=np.random.randon(n_h,n_x) #返回样本具有标准正态分布
    b1=np.zeros((n_h,1)) #注意维度
    W2=np.random.randn((n_y,n_h))*0.01
    b2=np.zeros((n_y,1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    return parameters
```
