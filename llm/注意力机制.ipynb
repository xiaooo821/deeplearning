# 原理介绍


## The Bottleneck Problem：   

  
在传统的序列到序列模型的编码阶段，源语句编码（source sentence encoding）会生成一个单一向量（single vector）。这个向量需要捕捉关于源语句的所有信息。但实际上，让一个向量完整且精准地包含长句子或者复杂语句中的全部语义、语法、语境等信息是非常困难的 。       
单一向量限制了encoder的表示能力。由于向量维度和存储信息能力的局限，它无法完美承载源语句的全部信息，就像一个狭窄的瓶口限制了水流的通过量一样，这就形成了信息瓶颈


**解决瓶颈问题的方法：attention**          
在解码器的每一步，注意力机制不是依赖单一的编码向量，而是聚焦于源序列的特定部分   

## Seq2Seq with Attention

这样的过程不断重复，decoder的每个隐藏状态都可以关注到encoder端不同位置的隐向量，从而预测出下一个单词。


## attention更通用定义

Attention 机制的核心思想是：**根据输入的不同部分的重要性，动态地分配不同的权重**。具体来说，Attention 通过对一组 value 向量进行加权平均，得到一个上下文相关的表示。这个加权平均的权重是通过 query 和 key 的相似度计算得到的。

通过attention机制，就可以得到一个综合了所有value信息的表示，同时纬度和value向量纬度也一致

好处： 
- 解决了信息瓶颈问题：
使得decoder端在每次生成新的隐藏状态的时候都可以关注到encoder端的所有位置向量信息。

- 解决了梯度消失问题：
因为每次生成都关注到了encoder端的所有信息，所以非常有效地缓解了梯度在RNN中传播过长而导致的梯度消失。
