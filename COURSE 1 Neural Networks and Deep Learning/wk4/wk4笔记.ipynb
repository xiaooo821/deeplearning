# 为什么深度这么有理？
- 通过一层一层的layers，deep learning能够通过初步的细节，一步一步提升其对数据的理解和表达
- 直观的，将神经网络的浅层看作简单的检测函数，如边缘检测，然后再网络的后一层，将它们组合起来，一边学习更多的功能。


# 为深层神经网络构建模块

- “cache” 记录来自正向传播单元的值并将其发送到反向传播单元，因为需要链式计算导数。

## 正向传播过程     
输入$a^[l-1]$  
输出 $ a^{l}$ ， 存储 $z^{[l]}$ (实际上也存储 $w^{[l]}, b^{[l]}$)      
传播公式如下：(用向量化表示)
$$Z^{[l]}=W^{[l}*A^{[l-1]}+b^{[l} $$
$$A^{[l]}=g^{[l]}(z^{[l]})$$

## 反向传播过程（？）
反向传播是计算损失函数 L 对网络中每个参数的偏导数。      
输入$da^{[l]}$   
输出 $da^{[l-1]},dW^{[l]},db^{[l]}$    
公式如下：
$$dz^{[l]}=da^{[l]}*g^{'[l]}(z^{[l]})$$
$$dw^{[l]}=dz^{[l]}*a^{[l-1]}$$
$$db^{[l]}=dz^{[l]}$$
$$da^{[l-1]}=(w^{[l]})^T*dz^{[l]}$$
