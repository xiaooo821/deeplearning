# 一些概念


- 在批量梯度下降中，我们每次使用一批（batch）样本来更新模型的参数。与随机梯度下降（SGD）每次只使用一个样本，或小批量梯度下降（Mini-batch Gradient Descent）每次使用一个小批量样本不同，批量梯度下降在整个训练集上计算损失和梯度。所以批量梯度下降要对梯度进行归一化操作/求均值 也就是所有梯度加和除以batch size


# 从0开始实现
- 在这一节中，我们将从零开始实现整个方法， 包括数据流水线、模型、损失函数和小批量随机梯度下降优化器。

```python
%matplotlib inline
import random
import torch
import d2l
```

生成数据集：  

我们根据一个已知w，b参数的线性数据集，加入噪音，然后用使用这个有限样本的数据集来恢复这个模型的参数。 


```python
def synthetic_data(w, b, num_examples):  #@save
    """生成y=Xw+b+噪声"""
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
# 注意，features中的每一行都包含一个二维数据样本， labels中的每一行都包含一维标签值（一个标量）。
```

```python
features.shape,labels.shape
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, (1)].numpy(), labels.numpy(), 1);
```

批量读取数据集：

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # 这些样本是随机读取的，没有特定的顺序
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
#在 Python 中，yield是一个关键字，用于定义生成器函数。生成器函数是一种特殊的函数，它返回一个迭代器，可以在迭代过程中逐个生成值，而不是一次性返回所有结果。
```

初始化模型参数

```python
w=torch.normal(0,0.01,size=(2,1),requires_grad=True)  #列向量
b=torch.zeros(1,requires_grad=True)
```

模型定义

```python
def linreg(X, w, b):  #@save
    """线性回归模型"""
    return torch.matmul(X, w) + b 
# torch.matmul是更通用的矩阵乘法函数。它可以处理多种维度的张量乘法。且支持广播机制
```

损失函数

```python
def squared_loss(y_hat, y):  #@save
    """均方损失"""
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

定义优化算法：    
用于更新操作 而不是计算梯度 所以使用torch.no_grad()。这是 PyTorch 中的一个上下文管理器。在这个上下文范围内，所有的计算都不会被记录用于求导。

```python
#params是所有参数 ，w和b的集合。直接在函数中进行更新～
def sgd(params, lr, batch_size):  #@save
    """小批量随机梯度下降"""
    with torch.no_grad():
        for param in params:  
            param -= lr * param.grad / batch_size  #最后再计算均值【很重要】
            param.grad.zero_()  #在更新完一个参数后，需要将其梯度清零
```

训练

```python
lr = 0.03
num_epochs = 3
batch_size=10
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = squared_loss(linreg(X, w, b), y)  # X和y的小批量损失
        # 因为l形状是(batch_size,1)，而不是一个标量。l中的所有元素被加到一起，
        # 并以此计算关于[w,b]的梯度
        l.sum().backward()
        sgd([w, b], lr, batch_size)  # 使用参数的梯度更新参数
    #在每个训练轮次结束时计算并打印出整个训练数据集上的损失值，以便监控模型的训练进展：
    with torch.no_grad(): 
        train_l = squared_loss( linreg(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

最后，来检验一下，实际与预测出来的w，b差值

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}')
```

注意，我们不应该想当然地认为我们能够完美地求解参数。 在机器学习中，我们通常不太关心恢复真正的参数，而更关心如何高度准确预测参数。 幸运的是，即使是在复杂的优化问题上，随机梯度下降通常也能找到非常好的解。 其中一个原因是，在深度网络中存在许多参数组合能够实现高度精确的预测。

# 简洁实现


```python
import numpy as np
import torch
from torch.utils import data
```

生成数据集  
- 这里features, labels分别是数据的输入和输出

```python
true_w=torch.tensor([2,-3.4]) #两个参数的向量
true_b=4.2
features, labels = synthetic_data(true_w, true_b, 1000) 
```

```python
# 读取数据集
# TensorDataset:用于将多个张量（Tensor）数据组合在一起，形成一个数据集（Dataset）
# DataLoader:可以帮助我们以小批量（mini - batch）的方式加载数据。
def load_array(data_arrays, batch_size, is_train=True):  #@save
    """构造一个PyTorch数据迭代器"""
    dataset = data.TensorDataset(*data_arrays)   #*是解包操作
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 10
data_iter = load_array((features, labels), batch_size)

# 定义模型
# Sequential类将多个层串联在一起。 当给定输入数据时，Sequential实例将数据传入到第一层， 然后将第一层的输出作为第二层的输入，以此类推
#在PyTorch中，全连接层在Linear类中定义。 我们将两个参数传递到nn.Linear中。 第一个指定输入特征形状，即2，第二个指定输出特征形状，输出特征形状为单个标量，因此为1。
from torch import nn   # nn是神经网络的缩写
net = nn.Sequential(nn.Linear(2, 1))

# 初始化模型参数
# 使用替换方法normal_和fill_来重写参数值。
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)

# 定义损失函数 
# 计算均方误差用MSELoss，也就是L2范数
loss = nn.MSELoss()

# 定义优化函数
##  net.parameters()返回一个迭代器，其中包含所有可训练的参数。
## SGD 就是随机梯度下降算法的实现函数
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

**训练**  
必备步骤：
- 前向传播：调用net生成预测值，并计算损失l 
- 反向传播：计算梯度
- 更新模型参数：调用优化器

```python
num_epoch=3
for epoch in range(num_epochs):
    for X,y in data_iter:
        l = loss(net(X),y)
        trainer.zero_grad() # 在进行反向传播计算梯度之前，需要先将模型参数的现有梯度清零。
        l.backward()
        trainer.step()  # 这个方法会根据之前计算出的梯度以及优化器中设置的学习率来更新模型的参数
    l=loss(net(features),labels)
    print(f'epoch {epoch + 1}, loss {l:f}')  #:f 表示将变量 l 按照浮点数的格式进行输出
```
