以上说明正向和反向都要满足，$\beta=0,\alpha=1$。也就意味着激活函数必须是$\sigma(x) = x$

通常神经网络的权重就是在零点附近，所以我们要满足在零点附近是恒等函数。也就是y(x)=x。我们发现，对于tanh(x)和relu(x)函数，在零点附近可以近似为线性函数y = x；而sigmoid(x)不行。
• 因此，我们对sigmoid(x)进行调整：
4 × sigmoid(x) - 2
