## 8.1 正交化（Orthogonalization）
正交化指两个变量成90度角。      
通过每次只调试一个参数，保持其它参数不变，而得到的模型某一性能改变是一种最常用的调参策略，我们称之为正交化方法（Orthogonalization）。
- 核心在于每次调试一个参数只会影响模型的某一个性能。这种方法能够让我们更快更有效地进行机器学习模型的调试和优化。

## 8.4 满足和优化指标
有时候，要把所有的性能指标都综合在一起，构成单值评价指标是比较困难的。       
解决办法是，我们可以把某些性能作为优化指标（Optimizing metic），寻求最优化值；而某些性能作为满意指标（Satisficing metic），只要满足阈值就行了。


有时候选择一个指标进行优化是合理策略，这个就是**优化指标**，而剩下的N-1就是**满足指标**，也就是满足某个阈值就可以了。一旦它们满足了阈值，就不用再费心优化了

## 8.5 训练 开发 测试集分布
机器学习的工作流程是，在训练集上训练不同模型，然后用开发集来评估不同的想法，并选择一个，保持创新提高模型在开发集上的表现性能。直到有一个好结果后，在测试集评估它。
- 如果dev sets和test sets不来自同一分布，那么我们从dev sets上选择的“最佳”模型往往不能够在test sets上表现得很好。

在之前的课程中我们已经介绍过，当样本数量不多（小于一万）的时候，通常将Train/dev/test sets的比例设为60%/20%/20%，   在没有dev sets的情况下，Train/test sets的比例设为70%/30%。【大数据时代，这种划分方式已经过时啦！】        
<font color='red'> 当样本数量很大（百万级别）的时候，通常将相应的比例设为98%/1%/1%或者99%/1%。
  

<font color='red'>另外，当发现评估指标无法对算法的优劣做出正确的排序时，那么需要考虑重新定义一个新的评估指标，以便更好地对分类器排序

## 8.8 为什么选择人类水平表现
贝叶斯最优误差/贝叶斯误差，是从x到y的最好理论函数，是永远无法超越的。


- 横坐标是训练时间，纵坐标是准确性。机器学习模型经过训练会不断接近human-level performance甚至超过它
- 但是，超过human-level performance之后，准确性会上升得比较缓慢，最终不断接近理想的最优情况，我们称之为bayes optimal error。


人类在很多任务上都表现得相当出色。所以，只要机器学习比人类表现差，你可以：

- 从人类获取标注数据：可以请人类对数据进行标注，以此作为训练数据来提升机器学习模型的性能。

- 通过手动错误分析获取建议：可以分析机器学习模型的错误，看看为什么人类能够正确处理这些任务。这有助于找出模型的问题所在。

\* 实际应用中，我们一般会用human-level error代表bayes optimal error。

- Online advertising

- Product recommendations

- Logistics(predicting transit time) 物流 预测从A地到B地的时间

- Loan approvals

① 这表明，提高性能的一个很好的途径是训练一个更大的网络，以降低4%的训练误差。你同意吗？

A. 【  】是的，因为有4%的训练误差表明你有很高的偏差。

B. 【  】是的，因为这表明你的模型的偏差高于方差。

C. 【  】不同意，因为方差高于偏差。

D. 【  】不同意，因为没有足够的信息，这什么也说明不了。

答案：

D.【 √ 】不同意，因为没有足够的信息，这什么也说明不了。

分析：<font color='blue'>
A 4%的训练误差不一定意味着很高的偏差。这取决于**任务的难度和目标精度**。如果目标误差非常低，那么4%可能是高偏差，但我们没有足够的信息来确定这一点。    
B **训练误差和验证误差接近，说明方差不高**。但这并不一定意味着偏差高，因为我们不知道目标误差是多少

人类表现 0.1%  
训练集误差 2.0%    
验证集误差 2.1%      
① 根据你的资料，以下四个选项中哪两个尝试起来是最有希望的？（两个选项）

A. 【  】尝试增加正则化。

D. 【  】训练一个更大的模型，试图在训练集上做得更好。

5. 你在测试集上评估你的模型，并找到以下内容：

人类表现 0.1%        
训练集误差 2.0%        
验证集误差 2.1%      
测试集误差 7.0%       
① 这意味着什么？（两个最佳选项）

A. 【  】你没有拟合验证集

B. 【  】你应该尝试获得更大的开发集

C. 【  】你应该得到一个更大的测试集

D. 【  】你对验证集过拟合了

答案：

B.【 √ 】你应该尝试获得更大的开发集

D.【 √ 】你对验证集过拟合了 

分析：
<font color='#DA3722'> 
B. 你应该尝试获得更大的开发集：如果开发集（验证集）过小，可能**无法代表整个数据分布，导致模型在实际测试集上表现不佳**。
增大开发集可以更好地评估模型的性能，确保模型不会在不同数据分布上表现出大的误差。          
    
D. 你对验证集过拟合了：测试集误差显著高于验证集误差，可能表明模型在验证集上表现良好但在测试集上泛化能力不足。              
     

ps：和上一题目的对比，上一题是训练集上面偏差大，可以通过正则项来处理。因为“正则化的作用是防止模型过拟合**训练数据**。减小正则化强度会使模型变得更复杂，从而可能更好地拟合训练数据”,注意是训练数据哦。验证集上面就和正则项没关系了。在这里测试集误差远高于训练集和验证集误差，表明模型在新数据上的泛化性能不佳。
