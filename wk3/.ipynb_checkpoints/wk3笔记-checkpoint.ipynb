# 激活函数

一定要使用激活函数，否则神经网络的输出就仅仅是输入函数的线性变化，因为线性函数的组合仍然是线性函数。

一般很少使用sigmoid，除非输出是y∈{0，1}，也就是二分类问题。其他情况tanh优化度更高，更普遍的是**用relu**



tanh
- 输出范围在-1到1之间，具有零中心性质，即均值为0，这有助于保持数据分布的对称性，减少了数据在传播过程中的偏移和不稳定性。因此，网络在学习过程中能够更容易地保持梯度的相对一致性，
- 缺点：如果输入很大或很小，函数的梯度就会很小，斜率接近于0，会减慢梯度下降的速度 

relu 
- 它收敛更快。线性意味着当x变大时，斜率不会平稳或“饱和” 。它没有其他激活函数具有的梯度消失问题（如Sigmoid或tanh）。

softmax
- 通常用于**多类分类的最后阶段**，将神经网络的原始输出转换为概率分布，表示每个类别的预测概率。并且概率总和为1
- 每个样本只能属于一个类别。


# 反向传播

首先要计算激活函数的导数
- sigmoid 
  $$a(x)=sigmoid(x)=\frac{1}{1+e^{-x}}$$
  $$a'(x)=a(x)(1-a(x))$$

- tanh 
$$ g(x)=tanh(x)=\frac{e^z-e^{-z}}{e^z+e^{-z}}$$
$$g'(x)=1-(g(x))^2$$

- relu
$$g(x)=max(0,x)$$
$$g'(x)=\begin{cases}
0 & if x<0\\
1 & if x>0\\
没有定义 & if x=0
\end{cases}$$

接着利用梯度更新参数
- 初始先给定一个w，沿着梯度下降的方向进行迭代，使更新后的L(w)损失函数不断变小.其中 $\eta$称为学习率或步长
 
 
 $$ w=w-\eta \frac{∂L(w)}{∂w}$$
             
通过迭代可以使得损失函数L(w)不断减小，直到为0

## 训练集、验证集、测试集
- 训练集:用于模型拟合的数据样本,即用于训练的样本集合,主要用来训练神经网络中的参数  


- 验证集：模型训练过程中单独留出的样本集，它可以用于**调整模型的超参数**和用于对模型的能力进行初步评估。
   - 如果没有设置验证集，我们通常得等到测试集才可以知道我们模型真正的实力，然后再来调整参数，这样子时间代价较高，通过验证集我们可以训练几个epoch后查看模型的训练效果及我们的网络是否出现异常，然后决定怎么调整我们的超参数。    
   
   
- 测试集：用来评估模最终模型的泛化能力。但不能作为调参、选择特征等算法相关的选择的依据。

2. 假设你已经建立了一个神经网络。您决定将权重和偏差初始化为零。以下哪项陈述是正确的？（选出所有正确项）        
 
A. 【 √  】第一隐藏层中的每个神经元将执行相同的计算。因此，即使在梯度下降的多次迭代之后，层中的每个神经元将执行与其他神经元相同的计算。        
分析：  

<font color=#A020F0 >
当将神经网络的权重和偏差初始化为零时，会导致网络中每一层的每个神经元的初始输出都相同。  
初始时，每个神经元在进行前向传播时会得到相同的输入信号，并使用相同的权重和偏置进行计算。这导致每个神经元的输出在初始阶段是相同的。 
在使用梯度下降等优化算法进行反向传播和权重更新时，神经网络试图通过调整权重来最小化损失函数。
如果所有神经元在初始时产生相同的输出，它们将接收到相同的梯度信号。这意味着它们的权重更新量也会是相同的，导致它们在整个训练过程中仍然会执行相同的计算。</font>

<font color=#A020F0 >这是为什么在实践中，神经网络的权重和偏置通常会**使用随机初始化**。</font>

## 维度关系的确定

首先是一些符号的表示

    
- $a^{[2]}_4$是第2层，第4个神经元的激活输出      


- $a^{[2]}$表示是第2层的激活向量   




上图中，
输入向量$X^{[0]}$,
- 维度：(2,1),2个输入特征，1个样本   

简单总结一下，这几个变量的维度是：   
W：行数是当前神经元个数，列数是前一层神经元个数（也就是输入层特征数）  
Z和b：行数是当前神经元的个数，列数是1  
X：行数是前一层神经元的个数，列数是1    



为什么X和b的是列数是1？
<font color=#A020F0 >
- 在前向传播过程中，神经网络是对每个样本逐个计算的。对于每个样本，**计算的输出是一个列向量**，即每个神经元对应一个标量输出，这些输出组成了z 向量。
- 而对于多个样本的计算，通常会将多个样本的列向量并在一起形成矩阵，**列数等于样本数量**。
</font >
